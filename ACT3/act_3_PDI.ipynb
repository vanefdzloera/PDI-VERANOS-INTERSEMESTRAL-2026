###**Parte II - Análisis de imagen en escala de grises en python**

**Objetivo**

En esta práctica se analiza el histograma de una imagen digital utilizando OpenCV y Matplotlib. Para ello, se carga una imagen, se convierte a escala de grises y posteriormente se obtiene su histograma con el fin de interpretar la distribución de intensidades de los píxeles y comprender cómo esta información describe las características visuales de la imagen.

**1. Cargar la imagen**

Como primer paso, se establece la conexión con Google Drive para acceder al archivo de imagen que será utilizado durante la práctica. Posteriormente, mediante la función `cv2.imread()` de OpenCV se carga la imagen denominada "ave.jpg", la cual servirá como base para realizar el análisis del histograma.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

img_dir = "/content/drive/MyDrive/IMAGENES/ave.jpg"

img = cv2.imread(img_dir)
cv2_imshow(img)

```

La imagen se cargó correctamente desde Google Drive y se visualizó utilizando la función `cv2_imshow()`. Esta imagen corresponde a un ave con un fondo desenfocado y será utilizada en las siguientes etapas del procesamiento digital de imágenes.

**2. Conversión a escala de grises**

Para facilitar el análisis del histograma, la imagen original se transforma a escala de grises mediante la función `cv2.cvtColor()`.

En este tipo de imágenes cada píxel posee únicamente un valor de intensidad entre 0 y 255, donde 0 representa el color negro y 255 el color blanco. Esta conversión elimina la información de color y permite estudiar únicamente la distribución de intensidades.

```python
img_gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gris)
```

Se obtuvo correctamente la versión en escala de grises de la imagen. En ella es posible apreciar únicamente los diferentes niveles de intensidad luminosa, conservando la información estructural de la imagen sin considerar el color.

**3. Cálculo del histograma**

Una vez obtenida la imagen en escala de grises, se calcula su histograma utilizando la función `cv2.calcHist()`. El histograma representa la distribución de los tonos de gris presentes en la imagen, mostrando cuántos píxeles existen para cada nivel de intensidad. Posteriormente, esta información se grafica mediante `plt.hist()` para facilitar su visualización e interpretación.

```python
hist = cv2.calcHist([img_gris],[0],None,[256],[0,256])


#Elaborando el Histograma
grises = np.arange(0,256)

plt.hist(grises,bins = 256, range = (0,256),weights = hist.ravel())
plt.title("Histograma escala de grises")
plt.ylabel("Cantidad de Pixeles")
plt.xlabel("Intensidad (0 = negro, 255 = Blanco)")
plt.show()
```

El histograma obtenido permite observar cómo se distribuyen los niveles de intensidad de la imagen. En el eje horizontal se representan los tonos de gris, donde 0 corresponde al negro y 255 al blanco, en el eje vertical se muestra la cantidad de píxeles que tienen cada uno de esos niveles de intensidad. Esta representación facilita el análisis del brillo, el contraste y la distribución de los tonos presentes en la imagen.
